```python
import sys
sys.path.append('..')

import numpy as np
from src.environment import NavigationEnv, ObstacleManager
import matplotlib.pyplot as plt
```

## Create Environment

```python
env = NavigationEnv(
    width=800,
    height=600,
    render_mode='rgb_array'
)

print(f"Action space: {env.action_space}")
print(f"Observation space: {env.observation_space}")
```

## Test Random Actions

```python
state, info = env.reset()
print(f"Initial state: {state}")
print(f"Info: {info}")

total_reward = 0
frames = []

for step in range(100):
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)
    
    frame = env.render()
    if frame is not None:
        frames.append(frame)
    
    total_reward += reward
    
    if terminated or truncated:
        print(f"Episode ended at step {step}")
        break

print(f"Total reward: {total_reward:.2f}")
print(f"Final distance: {info['distance_to_goal']:.2f}")
```

## Visualize Frames

```python
if frames:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    indices = np.linspace(0, len(frames)-1, 6, dtype=int)
    
    for idx, frame_idx in enumerate(indices):
        axes[idx].imshow(frames[frame_idx])
        axes[idx].set_title(f"Step {frame_idx}")
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()
```
